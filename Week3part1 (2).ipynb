<h1>Notebook to transform data in Wikipedia to a pandas dataframe</h1>


<h5>Let us start by importing pandas dataframe to process the data</h5>

```python
import pandas as pd
```

<h5>
<a href="https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M">This is url of Wikipedia page with required data which we will extract using read_html function </a> <br>


```python
url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
df_bor=pd.read_html(url)
df_b=df_bor[0]
df_b.head()
```

<h5> To remove rows with non-assigned boroughs </h5>

```python
q_df=df_b.query('Borough!="Not assigned"')
q_df.head()
```

<h5> To group neighbourhoods by common postal code and separate multiple neighbourhoods with same postal code by commas</h5>

```python
g_df =q_df.groupby(['Postal Code','Borough'],as_index=False).agg(lambda x: ", ".join(x))
g_df.head()
```

```python
for i in range (0,len(g_df)):
 if g_df.at[i,'Neighbourhood']=='Not assigned':
    g_df.at[i,'Neighbourhood']=g_df.at[i,'Borough']
```

```python
g_df.head()
```

<h5> To see if obtained table is accurate to the one to be obtained </h5>

```python
col_nam = ["Postal Code", "Borough", "Neighbourhood"]
check_frame = pd.DataFrame(columns=col_nam)

pc_list= ["M5G", "M2H", "M4B", "M1J", "M4G", "M4M", "M1R", "M9V", "M9L", "M5V", "M1B", "M5A"]

for pc in pc_list:
    check_frame = check_frame.append(g_df[g_df["Postal Code"]==pc], ignore_index=True)
    
check_frame
```
