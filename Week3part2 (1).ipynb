<h1>Notebook to get latitude and longitude of each neighbourhood </h1>

<h5>Let us start by importing pandas dataframe and csv file from last week containing needed data to be worked on this week,imported via read_csv</h5>

```python
import pandas as pd
neigh_df=pd.read_csv("Neighbourhoods.csv")
neigh_df.head()
```

<h5>
<a href="https://cocl.us/Geospatial_data">This is url of csv file containing co-ordinates of each postal code which is converted to another dataframe via read_csv </a> <br>


```python
loc_url="https://cocl.us/Geospatial_data"
loc_df=pd.read_csv(loc_url)
loc_df.head()
```

<h5> To merge dataframes based on common postal code </h5>

```python
geo_df=pd.merge(neigh_df,loc_df,on="Postal Code")
geo_df.head()
```

<h5> To check if final dataframe matches one specified in instructions </h5>

```python
col_nam = ["Postal Code", "Borough", "Neighbourhood","Latitude","Longitude"]
check_frame = pd.DataFrame(columns=col_nam)

pc_list= ["M5G", "M2H", "M4B", "M1J", "M4G", "M4M", "M1R", "M9V", "M9L", "M5V", "M1B", "M5A"]

for pc in pc_list:
    check_frame = check_frame.append(geo_df[geo_df["Postal Code"]==pc], ignore_index=True)
    
check_frame
```

<h5> To export above dataframe as csv file for ease of use next week </h5>

```python
geo_df.to_csv("Coordinates.csv",index=False)
```
